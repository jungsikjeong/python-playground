```python
# pip install openpyxl

import pandas as pd

raw= pd.read_excel('product.xlsx',engine='openpyxl')

# 특정 컬럼에 있는 데이터를 가지고 새로운 컬럼을 만들 수 있는 방법

# 방법1 
raw['부가세포함'] =raw['판매가'] * 1.1 

# 방법2 함수 사용
def 함수(data):
  return data * 1.1
raw['부가세포함']=raw['판매가'].apply(함수)

# 특정 컬럼에 있는 데이터를 가지고 새로운 컬럼을(카테고리) 만들 수 있는 방법
# 방법 3 '정규식'(regex) 사용

# 'abc'가 'abcdef'에 들어있는지 검사해주는 정규식
# a = re.search('abc','abcdef')
# print(a) # print: <re.Match object; span=(0, 3), match='abc'>

import re

def 카테고리목록생성함수(data):
  # 어 .. 만약에 상품목록이 7102 Metal Chair면 '의자'를 리턴하기
  if(re.search('Chair',str(data))):
    return '의자'
  if (re.search('Table',str(data))):
    return '테이블'
  

raw['카테고리'] = raw['상품목록'].apply(카테고리목록생성함수)

raw
```
